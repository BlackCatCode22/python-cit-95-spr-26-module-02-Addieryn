#### Exercise 1: Write a program which repeatedly reads integers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the integers. If the user enters anything other than an integer, detect their mistake using try and except and print an error message and skip to the next integers.

```python
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

````python
import numpy as np
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input")

average = sum(numbers) / len(numbers)

if numbers:
    print(sum(numbers), len(numbers), average)
        
````

#### Exercise 2: Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

````python
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input")
    
if numbers:
    print("Maximum:", max(numbers))
    print("Minimum", min(numbers))
else:
    print("No numbers were entered.")
````
