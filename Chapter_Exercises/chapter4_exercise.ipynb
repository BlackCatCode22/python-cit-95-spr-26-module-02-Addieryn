#### Exercise 6: Rewrite your pay computation with time-and-a-half for overtime and create a function called computepay which takes two parameters (hours and rate).
```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0

````python
def compute_pay(hours, rate):
    reg_hours = 40

    my_hours = float(hours)
    my_rate = float(rate)
    
    if my_hours > reg_hours:
        overtime = my_hours - reg_hours
        over_rate = my_rate * 1.5

        pay = round((reg_hours * my_rate) + (overtime * over_rate), 2)

        print(f'Enter Hours: {my_hours}')
        print(f'Enter Rate: {my_rate}')
        print(f'Pay: {pay}')
    else:
        pay = round(my_hours * rate, 2)

        print(f'Enter Hours: {my_hours}')
        print(f'Enter Rate: {my_rate}')
        print(f'Pay: {pay}')

try:
    compute_pay(45, 10)
    
except NameError:
    print('Error, please enter numeric input')
````

Run the program repeatedly to test the various different values for input.

````python
def compute_grade(my_score):
    score = float(my_score)
    if score < 0 or score > 1:
        print('Bad score')
    else:  
        if score >= 0.9:
            print('A')
        elif score >= 0.8:
            print('B')
        elif score >= 0.7:
            print('C')
        elif score >= 0.6:
            print('D')
        else:
            print('F')
            

try:        
    compute_grade(.77)
except NameError:
    print('Bad score')
````
